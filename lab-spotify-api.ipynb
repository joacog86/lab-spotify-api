````python
import sys
from config import *

import numpy as np
import pandas as pd
import time

import spotipy
import json
from spotipy.oauth2 import SpotifyClientCredentials

#Initialize SpotiPy with user credentias #
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=Client_ID,
                                                           client_secret=Client_Secret))
````

````python
#Loading the DataBase of Not Hot 100

hot100 = pd.read_csv('hot100.csv')
nothot100 = pd.read_csv('nothot100_slim.csv')
````

````python
hot100['H_or_N'] = 'H'
nothot100['H_or_N'] = 'N'
````

````python
#Function giving a title and an artist an a limit 

def search_song(title, artist, limit=1):
    list_of_ids = []

    try:
        song_name = title
        artist_name = artist
        query = f" {song_name} song_name  {artist_name} artist_name "

        results = sp.search(q=query, limit=1)
        id_song = results['tracks']['items'][0]['uri']
        list_of_ids.append(id_song)

    except:
      print("Song not found!")
      list_of_ids.append("")

    return list_of_ids

# Code to remove songs without IDs from the databases.
````

````python
def retrieve_song_id(df, chunk_size=50, sleep_time=35):
    """
    Process chunks of a DataFrame using a specified function and concatenate the results.

    Parameters:
    - df: DataFrame
    - chunk_size: int, optional (default=50)
    - sleep_time: int, optional (default=35)

    Returns:
    - final_df: DataFrame
    """
    num_chunks = int(np.ceil(len(df) / chunk_size))

    # Create multiple lists of DataFrames
    list_of_chunks = []

    for i in range(num_chunks):
        start_idx = i * chunk_size
        end_idx = (i + 1) * chunk_size
        chunk_df = df.iloc[start_idx:end_idx]
        
        # Apply the search_song_function to each row of the chunk
        chunk_df['id'] = chunk_df.apply(lambda row: search_song(row['song'], row['artist'])[0], axis=1)
        
        time.sleep(sleep_time)
        print(f"Chunk {i + 1}:\n{chunk_df}\n")
        list_of_chunks.append(chunk_df)

    # Concatenate the list of DataFrames into a new DataFrame
    final_df = pd.concat(list_of_chunks, ignore_index=True)

    return final_df
````

````python
hot100_id = retrieve_song_id(hot100)
````

````python
# Function to retrieve audio features for a list of track IDs
def get_audio_features(track_ids):
    try:
        audio_features = sp.audio_features(track_ids)
        return audio_features
    except Exception as e:
        print(f"Error fetching audio features: {e}")
        return []
````

````python
def retrieve_audio_features(df, chunk_size=50, sleep_time=30):
    """
    Retrieve audio features for chunks of a DataFrame and concatenate the results.

    Returns:
    - final_df: DataFrame
    """
    num_chunks = int(np.ceil(len(df) / chunk_size))

    # Create a list to store audio features DataFrames
    audio_features_list = []

    for i in range(num_chunks):
        start_idx = i * chunk_size
        end_idx = (i + 1) * chunk_size
        chunk_df = df.iloc[start_idx:end_idx]
        
        # Retrieve audio features for the chunk using get_features_function
        track_ids = chunk_df['id'].tolist()
        
        try:
            audio_features = get_audio_features(track_ids)
            
            # Check if audio_features is not empty
            if audio_features:
                audio_features_df = pd.DataFrame(audio_features)
                audio_features_list.append(audio_features_df)
            else:
                print(f"No audio features retrieved for chunk {i + 1}.")
                
        except Exception as e:
            print(f"Error fetching audio features for chunk {i + 1}: {e}")
        
        time.sleep(sleep_time)
        print(f"Audio features retrieved for chunk {i + 1}.\n")

    # Concatenate the list of audio features DataFrames into a single DataFrame
    final_df = pd.concat(audio_features_list, ignore_index=True)

    return final_df
````

````python
hot100_audio_features = retrieve_audio_features(hot100_id)
````

````python
#Concatenating list of songs + audio features

hot100_total_df = pd.concat([hot100_id, hot100_audio_features], axis=1)
````
